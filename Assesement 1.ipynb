### Retrieve all cars for a specified cylinder number

### Retrieve a specific number of columns (2 or up to 5 columns) related to an  individual car by id 

For this, take `car_id` as input and then used a function `random_colums` to generate each time the data with desired number of columns 

```python
car_id = int(input("Enter Car Id (should be in range from 1 to {}):  ".format(car_sales_america.shape[0])))
```

```python
no_of_columns=int(input("Enter number of columns to query for car_ID {} (range: 2-5): ".format(car_id)))
```

```python
def random_colums():
    return car_sales_america.sample(n=no_of_columns,axis='columns')[car_sales_america.car_ID == car_id ]
```

```python
specific_number_of_columns=random_colums()
```

# Utilising data from a data frame

Now, the system should allow the user to utilise the data to perform the following  queries:
- Retrieve the car names alphabetically 
- Retrieve summary of sales (total car price) for each car body 
- Retrieve the top 5 car sale by price (the most expensive) and per car body

### Retrieve the car names alphabetically  

```python
only_car_names=car_sales_america.sort_values(by="CarName").CarName
```

```python
only_car_names
```

###  Retrieve summary of sales (total car price) for each car body

For this we have to group data according to similar carbody and then have to add price of each car with similar body type

```python
car_sales=car_sales_america.groupby('carbody')['price'].sum()
```

```python
car_sales
```

### Retrieve the top 5 car sale by price (the most expensive) and per car body

Similarly, we have to group data first acc to body type and then with cars of similar body types. From than data we have to extract the first 5 car prices, sorted in decending order

```python
car_sales=car_sales_america.groupby(['carbody','CarName'])[['price']].sum('price').head(5)
```

```python
car_sales.sort_values('price', ascending=False)
```

# Visualise data from a data frame

Now, the system should allow the user to visualise the data as follows :
- Display the number of cars per fuel system using a bar chart 
- Display the horsepower of the top 5 car sale by price (the cheapest) using a  subplot
- Display the selection of data by your choice to present the customers’ buying  behaviour using a suitable visualisation

### Display the number of cars per fuel system using a bar chart  
Group the data with similar `fuelsystem` and make a column to have the count and simply pass the datafram in `plot` function

```python
fuel_count=car_sales_america.groupby('fuelsystem')['fuelsystem'].count()
```

```python
fuel_count.plot(kind='bar');
```

### Display the horsepower of the top 5 car sale by price (the cheapest) using a  subplot
First we have to group data by `CarName` and according extract the `horsepower` of least expensive cars

```python
horsepower=car_sales_america.groupby('CarName',as_index=False)[['CarName','price','horsepower']].sum('price').sort_values('price').head(5)
```

```python
horsepower
```

```python
plt.subplot(1,2,1)
plt.xticks(rotation=45, ha='right')
plt.bar(horsepower.CarName,horsepower.horsepower)
plt.show()
```
